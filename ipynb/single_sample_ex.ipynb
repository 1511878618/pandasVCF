### This example notebook describes simple usage of pdVCFsingle, a module for parsing VCF files using the pandas library.  The single indicates that it does not load multiple samples into the dataframe, but can parse a single sample from a multi-sample VCF.  

# Libraries

```python
#Import pdVCFsingle package
import matplotlib as plt
import sys
sys.path.append( '../src/single_sample/' )
from pdVCFsingle import *
```

# Example File Path

```python
vcf_path = vcf_path = '../test_data/NA128_17member_pedigree_10k.vcf.gz'
```

# Creating Vcf object

### Initiate Vcf object by specifying the sample_id string and the columns the user wants to include for parsing.  Only the CHROM, POS, REF, ALT, and FORMAT fields are required. A parser for the INFO field is in development.  Some VCF files are quite large and will not fit in memory, therefore the user can specify the chunksize which allows iteration through the VCF.  A chunksize of 100,000 requires ~10seconds to parse.

```python
vcf_chunk = Vcf(vcf_path, sample_id='NA12878', cols=['#CHROM', 'POS', 'REF', 'ALT','FORMAT'], chunksize=1000)
```

### This function call will return the vcf chunks, and initiate the pandas df.

```python
vcf_chunk.get_vcf_df_chunk()
```

### The following attributes have been created for the vcf object

```python
dir(vcf_chunk)[19:]
```

### The vcf dataframe has the user-specified columns

```python
vcf_chunk.df.head(10)
```

```python
vcf_chunk.sample_id in vcf_chunk.df.columns
```

### Variant annotations are often useful and several can be generated using add_variant_annotations to create a df_annot object, please see the add_variant_annotations doc for additional details.

```python
vcf_chunk.add_variant_annotations(split_columns={'AD':2}, inplace=False, verbose=True)
vcf_chunk.df_annot.head(10)
```

#### The df object is still available in the original format

```python
vcf_chunk.df.head()
```

### If the user wants to conserve memory the inplace argument will parse the df and drop any redunant columns

```python
vcf_chunk.add_variant_annotations(split_columns={'AD':2}, inplace=True, verbose=True)
vcf_chunk.df.head(10)
```

### Note: add_variant_annotations inplace=True can only be run once as the FORMAT and sample genotype column is dropped

```python
vcf_chunk.add_variant_annotations(split_columns={'AD':2}, inplace=True, verbose=True)
```
